```python
import requests
from bs4 import BeautifulSoup

url = 'https://indeks.kompas.com/terpopuler'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'
}

print(f"Mengambil data dari: {url}")

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, 'html.parser')

    article_container = soup.find('div', class_='articleList -list')

    if article_container:
        articles = article_container.find_all('div', class_='articleItem')
        
        print(f"\n--- BERITA TERPOPULER DARI KOMPAS.COM ---\n")
        
        for index, article in enumerate(articles, 1):
            title_element = article.find('h2', class_='articleTitle')
            link_element = article.find('a', class_='article-link')
            category_element = article.find('div', class_='articlePost-subtitle')
            date_element = article.find('div', class_='articlePost-date')
            
            title = title_element.get_text(strip=True) if title_element else 'Judul tidak ditemukan'
            link = link_element.get('href', '') if link_element else 'Link tidak ditemukan'
            category = category_element.get_text(strip=True) if category_element else 'Kategori tidak ditemukan'
            date = date_element.get_text(strip=True) if date_element else 'Tanggal tidak ditemukan'
            
            print(f'{index}. Judul: "{title}"')
            print(f"   Kategori: {category}")
            print(f"   Tanggal: {date}")
            print(f"   Link: {link}\n")
            
    else:
        print("Wadah artikel ('articleList -list') tidak ditemukan. Struktur halaman mungkin telah berubah.")

except requests.exceptions.RequestException as e:
    print(f"Gagal terhubung ke situs: {e}")
```
