```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.tempo.co'

print(f"Mengambil data dari {url} ...")

try:
    response = requests.get(url)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, 'html.parser')

    trending_header = soup.find('span', string='ARTIKEL TRENDING')

    if trending_header:
        main_container = trending_header.find_parent('div', class_='flex-col')
        articles = main_container.find_all('figure')
        
        print(f"\n--- ARTIKEL TRENDING SAAT INI DI TEMPO.CO ---\n")
        
        for index, article in enumerate(articles[1:], 1):
            figcaption = article.find('figcaption')
            
            if figcaption:
                link_element = figcaption.find('a')
                
                if link_element:
                    title = link_element.get_text(strip=True)
                    
                    link_url = link_element.get('href', '')
                    
                    full_link = url + link_url if link_url.startswith('/') else link_url

                    print(f'{index}. Judul: "{title}"')
                    print(f"   Link: {full_link}\n")

    else:
        print("Tidak dapat menemukan kotak 'ARTIKEL TRENDING'. Struktur halaman mungkin telah berubah.")

except requests.exceptions.RequestException as e:
    print(f"Gagal terhubung ke situs: {e}") 
```

```python
import requests
from bs4 import BeautifulSoup
import re

url = 'https://www.tempo.co'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'
}

print(f"Mengambil data dari {url}...")

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, 'html.parser')

    trending_header = soup.find('div', string=re.compile(r'TOPIK TRENDING'))

    if trending_header:
        topic_list_ul = trending_header.find_next_sibling('ul')
        
        if topic_list_ul:
            topics = topic_list_ul.find_all('li')
            
            print("\n--- TOPIK TRENDING SAAT INI DI TEMPO.CO ---\n")
            
            for index, topic_item in enumerate(topics, 1):
                link_element = topic_item.find('a')
                
                if link_element:
                    topic_name = link_element.get_text(strip=True)
                    link_url = link_element.get('href', '')
                    full_link = url + link_url if link_url.startswith('/') else link_url

                    print(f'{index}. Topik: "{topic_name}"')
                    print(f"   Link: {full_link}\n")
        else:
            print("List topik (<ul>) tidak ditemukan setelah header. Struktur mungkin berubah.")
            
    else:
        print("Header 'TOPIK TRENDING' tetap tidak ditemukan. Mohon pastikan kembali di 'View Page Source' (Ctrl+U).")

except requests.exceptions.RequestException as e:
    print(f"Gagal terhubung ke situs: {e}")
```
